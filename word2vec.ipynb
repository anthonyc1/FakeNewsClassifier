```python
from nltk.tokenize import sent_tokenize, word_tokenize 
import warnings 

warnings.filterwarnings(action = 'ignore') 
  
import gensim 
from gensim.models import Word2Vec
import pandas as pd
import numpy as np
```

```python
# load the model for google word2vec

# for each document, find the word vectors
# then sum/avg function

# return a csv
# dimensions: x - num of docs, y - size of word embedding + 1 col for Real/Fake
# will use for x,y test/train

def loadVectors(vectorPath):
    '''
    Load vectors from given vectorPath
    '''
    
    import gensim
    # Load Google's pre-trained Word2Vec model vector.
    print('loading the vector takes about 1.5 mins...')
    model =gensim.models.KeyedVectors.load_word2vec_format(vectorPath, binary=True) 
    print('vector loaded')
    
    #remove code below and return the model if you wish to train the model more
    word_vectors = model.wv
    del model
    
    return word_vectors
```

```python
# obtain Google's pretrained word_vectors
word_vectors = loadVectors('./model/GoogleNews-vectors-negative300.bin')
```

```python
# obtain col from df
content = df['content_new']

# initialize blank array of size 1x300
blank = [0] * 300
blank = np.array(blank)
blank
```

```python
wv = []
index = 0
for article in content:
    try:
        words = article.split(' ')
        vector_sum = 0
        count = 0
        for word in words:
            try:
                vector_sum += word_vectors.word_vec(word)
                count+=1
            except:
                continue
        try:
            avg_vector = vector_sum/count
#             avg_vector.append(target[index])
            wv.append(avg_vector)
        except:
            wv.append(blank)
    except:
        wv.append(blank)
    index+=1
```

```python
# get a dataframe that is number of articles x 300
df2 = pd.DataFrame(wv)
```
